```python
##IGNORE THIS PART FOR NOW!!!!!!!!! THE CODE STARTS BELOW AS INDICATED
#local refinement
#c: Top elitist chromosome in the Population
#h: number of fitness function spent by local refinement or genetic search
#v: Maximum number of allowed fitness function calls(FFC)
#u: Number of spent FFCs, u = m+2*h*t; t is number of generation, m is size of population
#F: Total number of features
def localRefine(h,c,v,u,F):
    T = np.zeros((h,F))
    Sc = c.count(1)            #number of features in the elitist chromosome
    Ti_fit = [0]*h
    #Generate h duplicates of c and do local refinement for each of them
    for i in range(h):
        T[i] = c
        #b is the remaining FFCs ratio
        b = (v-u)/v
        #p and q have the same max limit determined by the remaining FFC 
        p = round(min(Sc,F-Sc)*b*random.random())
        #add p features to T[i]
        for j in range(p):
            ADD(T[i])
        q = round(min(Sc,F-Sc)*b*random.random())
        #delete q features from T[i]
        for j in range(q):
            DEL(T[i])
        Ti_fit[i] = getfit(T[i])[2]
    Tstar = max(Ti_fit)
    Tstar_position = Ti_fit.index(Tstar)
    #Replace the elitist chromosome c if the refined one has better fit
    if Tstar > getfit(c)[2]:
        c = T[Tstar_position]
    return c

        
        
        
    
```

```python
def genpop(data, pop):  
    i = np.where(pop==0)
    new_data = data.copy()
    new_data.drop(new_data.columns[i], axis=1, inplace=True)
    return new_data
```

```python
# find fitness for selection
# do ML method
def ML(x_train,x_test):
    # kernel='rbf'时（default），为高斯核，gamma值越小，分类界面越连续；gamma值越大，分类界面越“散”，分类效果越好，但有可能会过拟合
    # decision_function_shape='ovr'时，为one v rest，即一个类别与其他类别进行划分，
    clf = svm.SVC(C=0.8, kernel='rbf', gamma=20, decision_function_shape='ovr')
#     clf = MLPClassifier(solver='lbfgs',hidden_layer_sizes = (1000,1000))
#     clf = RandomForestClassifier()
    clf.fit(x_train, Y_train)
    fitness = clf.score(x_test, Y_test)
    #y_hat = clf.predict(x_test)
    #show_accuracy(y_hat, Y_test)
    return fitness
```

```python
# get fitness value of  population
def getfit(population):
    m = len(population)
    accuracy = np.zeros(m)
    for i in range(m):
        train_S = genpop(X_train, population[i]) # representing ith train dataset with selected feature
        test_S = genpop(X_test, population[i]) # representing ith test dataset with selected feature
        accuracy[i] = ML(train_S, test_S)
    return accuracy
```

```python
# mating process (genes crossover)
def crossover(parent1, parent2):
    if np.random.rand() < cross_rate:
        i = np.random.randint(0, DNA_size)                          # select another individual from population
        child1 = np.append(parent1[0:i+1],parent2[i+1:])  
        return child1
    else:
        return parent1
```

```python
#We could discuss if this is useful. From this test run, it seems that mutation could not affect the result that much
# mutation for child
def mutation(child):
    if np.random.rand() < mutation_rate:
        point = np.random.randint(0, DNA_size)
        child[point] = 1 if child[point] == 0 else 0
    return child 
```

```python
#convert a list of chromosome string to a 2d array 
def getchrom(liststr):
    n = len(liststr)
    result = np.zeros((n, DNA_size))
    for i in range(n):
        result[i] = np.array([int(d) for d in str(liststr[i])])
    return result
```
